```python
# Building the model with the example illustrated in class
# Quadratic equation: aX^2 + bX + c
a = 0
b = 0
c = 0

L = 1e-4 # learning rate
epochs = 10000 # number of epochs

n = float(len(X))

for i in range(epochs):
    Y_pred = a*(X**2) + b*X + c # the current prediction of Y
    D_a = (-2/n) * sum(X**2 * (Y-Y_pred)) # derivative wrt a
    D_b = (-2/n) * sum(X * (Y-Y_pred)) # derivative of loss function wrt b
    D_c = (-2/n) * sum(Y-Y_pred) # derivative of loss function wrt c
    a -= L*D_a
    b -= L*D_b
    c -= L*D_c
    print(f'Epoch {i}: {a} {b} {c}')
```
